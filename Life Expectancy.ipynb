```python
#Data manipulation

#Drop Year Column
Life_Expectancy = Life_Expectancy.drop('Year', axis = 1)
```

```python
#Handling Missing Data
#Delete Empty rows where Life Expectancy is blank
Life_Expectancy = Life_Expectancy.dropna(subset=['Life_expectancy'], axis = 0)
```

```python
Life_Expectancy.fillna(0, inplace = True)
```

```python
#No missing values
sns.heatmap(Life_Expectancy.isnull())
```

```python
#Convert Categorical Data into Numerical Data using Get Dummies (One Hot Encoding)

Life_Expectancy = pd.get_dummies(Life_Expectancy, columns = ['Status'])
```

```python
Life_Expectancy = pd.get_dummies(Life_Expectancy, columns = ['Country'])
```

```python
#Linear Regression

from sklearn import linear_model
from sklearn.model_selection import train_test_split

x = Life_Expectancy.drop(['Life_expectancy'], axis =1)
y = Life_Expectancy.Life_expectancy

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.4, random_state = 1)
```

```python
#Create an instance and fit the model
reg = linear_model.LinearRegression()

#Fitting the train and test data
reg.fit(x_train, y_train)

print(reg.score(x_test, y_test))
```

```python
#Predictions

y_pred = reg.predict(x_test)
```
